```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from numpy import asarray
```

```python
# Define 4 by 4 original patch
orig_4_4 = np.array([[1, 3, 5,3 ], [6,2,0,8], [4,6,1,4], [2,8,0,3]])
print(orig_4_4)
```

```python
def subsample(x_in):
    h_in, w_in = x_in.shape
    h_out, w_out = int(np.ceil(h_in / 2)), int(np.ceil(w_in / 2))
    x_out = np.zeros((h_out, w_out))

    for i in range(h_out):
        for j in range(w_out):
            h_start = i * 2
            w_start = j * 2
            patch = x_in[h_start:h_start+2, w_start:w_start+2]
            x_out[i, j] = np.mean(patch)
    return x_out
```

```python
print("Original:")
print(orig_4_4)
print("Subsampled:")
print(subsample(orig_4_4))
```

```python
def maxpool(x_in):
    h_out = int(np.floor(x_in.shape[0] / 2))
    w_out = int(np.floor(x_in.shape[1] / 2))
    x_out = np.zeros((h_out, w_out))

    for i in range(h_out):
        for j in range(w_out):
            h_start = i * 2
            w_start = j * 2
            patch = x_in[h_start:h_start+2, w_start:w_start+2]
            x_out[i, j] = np.max(patch)

    return x_out
```

```python
# Define 2 by 2 original patch
orig_2_2 = np.array([[6, 8], [8,4]])
print(orig_2_2)
```

```python
def duplicate(x_in):
    h_out = x_in.shape[0] * 2
    w_out = x_in.shape[1] * 2
    x_out = np.zeros((h_out, w_out))

    for i in range(x_in.shape[0]):
        for j in range(x_in.shape[1]):
            x_out[2*i:2*i+2, 2*j:2*j+2] = x_in[i, j]

    return x_out
```

```python
print("Original:")
print(orig_2_2)
print("Duplicated:")
print(duplicate(orig_2_2))
```

```python
def max_unpool(x_in, x_high_res):
    h_out, w_out = x_in.shape[0] * 2, x_in.shape[1] * 2
    x_out = np.zeros((h_out, w_out))

    for i in range(x_in.shape[0]):
        for j in range(x_in.shape[1]):
            h_start = i * 2
            w_start = j * 2
            patch = x_high_res[h_start:h_start+2, w_start:w_start+2]
            max_index = np.unravel_index(np.argmax(patch), (2, 2))
            x_out[h_start + max_index[0], w_start + max_index[1]] = x_in[i, j]

    return x_out
```

Finally, we'll try upsampling using bilinear interpolation.  We'll treat the positions off the image as zeros by padding the original image and round fractional values upwards using np.ceil()

```python
def bilinear(x_in):
    h_out, w_out = x_in.shape[0] * 2, x_in.shape[1] * 2
    x_out = np.zeros((h_out, w_out))

    # Padding the input image to avoid index issues
    x_in_pad = np.zeros((x_in.shape[0] + 1, x_in.shape[1] + 1))
    x_in_pad[0:x_in.shape[0], 0:x_in.shape[1]] = x_in

    for i in range(x_in.shape[0]):
        for j in range(x_in.shape[1]):
            # Calculate the position of the original value in the output
            x_out[2*i, 2*j] = x_in[i, j]  # original pixel value

            # Bilinear interpolation for horizontal and vertical neighbors
            if i < x_in.shape[0] - 1:
                x_out[2*i+1, 2*j] = (x_in[i, j] + x_in[i+1, j]) / 2
            if j < x_in.shape[1] - 1:
                x_out[2*i, 2*j+1] = (x_in[i, j] + x_in[i, j+1]) / 2
            if i < x_in.shape[0] - 1 and j < x_in.shape[1] - 1:
                x_out[2*i+1, 2*j+1] = (x_in[i, j] + x_in[i+1, j] + x_in[i, j+1] + x_in[i+1, j+1]) / 4

    return x_out
```
